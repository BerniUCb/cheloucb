## Filtrado con Valores Faltantes
- El método `filter` excluye los valores `null` por defecto.
- Polars tiene una familia de métodos de comparación que terminan en `_missing` que incluyen valores nulos.
- Por ejemplo, el método `eq_missing` compara la igualdad de un `null` con el argumento.

```python
coffees = pl.read_csv("coffee_sales.csv", try_parse_dates=True).sort("time_of_purchase")
coffees.head(2)
```

```python
coffees.null_count()
```

```python
coffees.filter(pl.col("coffee_name") != "Mocha")
# coffees.filter(pl.col("coffee_name").ne("Mocha"))
```

```python
# Conserva todo lo que no sea explícitamente "Mocha", incluyendo los valores faltantes.
coffees.filter(pl.col("coffee_name").ne_missing("Mocha"))
```

- Podemos pasar `None` para comparar la igualdad con valores `null`/faltantes...

```python
# coffees.filter(pl.col("coffee_name").eq(None)) # == None, no funciona
coffees.filter(pl.col("coffee_name").eq_missing(None))
```

- ...pero la solución más idiomática es el método `is_null`.
- El método `is_null` devuelve una `Series` booleana donde `true` representa un valor faltante.

```python
coffees.filter(pl.col("coffee_name").is_null()) # is None
```

```python
coffees.filter(pl.col("coffee_name").is_not_null()) # not is None
```

## Aplicando Lógica AND
- El operador `&` aplica lógica AND entre dos expresiones booleanas.
- Ambas condiciones deben cumplirse para que la evaluación sea verdadera.
- La lógica AND (`&`) se llama "conjunción" en matemáticas.

```python
coffees = pl.read_csv("coffee_sales.csv", try_parse_dates=True).sort("time_of_purchase")
coffees.head(2)
```

- Envuelve cada expresión en paréntesis para aislarla. De lo contrario, Polars se confundirá.
- El siguiente ejemplo extrae las filas con `milk_type` "Soy" y `size` mayor que 18.

```python
coffees.filter((pl.col("milk_type") == "Soy") & (pl.col("size") > 18))
```

- Los métodos permiten omitir los paréntesis.
- Alternativamente, asigna las expresiones a variables.

```python
# coffees.filter((pl.col("milk_type") == "Soy") & (pl.col("size") > 18))
# coffees.filter(pl.col("milk_type").eq("Soy") & pl.col("size").gt(18))

with_soy_milk = pl.col("milk_type") == "Soy"
with_size_greater_than_18 = pl.col("size") > 18

coffees.filter(with_soy_milk & with_size_greater_than_18)
```

## Aplicando Lógica OR
- El operador `|` aplica lógica OR entre dos expresiones booleanas.
- Al menos una condición debe ser verdadera para que la evaluación final sea verdadera.

```python
coffees = pl.read_csv("coffee_sales.csv", try_parse_dates=True).sort("time_of_purchase")
coffees.head(2)
```

- El siguiente ejemplo filtra las filas donde `price` es menor o igual a 3 o `location` es igual a "Downtown".

```python
coffees.filter((pl.col("price") <= 3) | (pl.col("location") == "Downtown"))
```

- El método `le` es equivalente al símbolo `<=`.
- Los métodos permiten omitir los paréntesis.

```python
# coffees.filter(pl.col("price").le(3) | pl.col("location").eq("Downtown"))
```

- Para simplificar el código, asigna las expresiones a variables descriptivas.

```python
cheap_coffee = pl.col("price") <= 3
in_downtown = pl.col("location") == "Downtown"
coffees.filter(cheap_coffee | in_downtown)
```

## Precedencia de Operadores
- Polars prioriza el operador AND (`&`) sobre el operador OR (`|`).

```python
coffees = pl.read_csv("coffee_sales.csv", try_parse_dates=True).sort("time_of_purchase")
coffees.head(2)
```

- Comencemos definiendo algunas expresiones booleanas.

```python
is_almond_milk = pl.col("milk_type") == "Almond"
is_iced = pl.col("iced")
is_cheap = pl.col("price") <= 4
```

- ¿Cómo interpreta Polars el filtro de abajo?
- Opción A ❌: Extraer filas que sean (leche de almendra O helado) Y barato
- Opción B ✅: Extraer filas que sean leche de almendra O (helado Y barato)
- El `&` prevalece, así que Polars agrupa `is_iced & is_cheap` juntos.

```python
coffees.filter(is_almond_milk | is_iced & is_cheap)
```

- Usa paréntesis para agrupar las expresiones relacionadas.
- Usa paréntesis incluso si la inferencia por defecto de Polars es correcta.

```python
# Precedencia explícita es mejor
coffees.filter(is_almond_milk | (is_iced & is_cheap))
```

## OR Exclusivo
- El operador OR exclusivo (`^`) requiere que solo se cumpla una condición.
- Si una condición es verdadera, la otra debe ser falsa.
- Si ambas condiciones son verdaderas, la expresión `^` se evalúa como falsa.


```python
coffees = pl.read_csv("coffee_sales.csv", try_parse_dates=True).sort("time_of_purchase")
coffees.head(2)
```

- Filtremos las filas con ubicación "Uptown" o leche "Oat".
- Polars _excluirá_ una fila si tiene tanto ubicación "Uptown" como leche "Oat".

```python
uptown = pl.col("location") == "Uptown"
oat_milk = pl.col("milk_type") == "Oat"

# coffees.filter(uptown | oat_milk)
coffees.filter(uptown ^ oat_milk)
```

## Filtrado de Valores Únicos y Duplicados

```python
import polars as pl
coffees = pl.read_csv("coffee_sales.csv", try_parse_dates=True).sort("time_of_purchase")
coffees.head(2)
```

- El método `is_unique` identifica si un valor de fila es único (aparece solo una vez en la columna).
- El método `is_duplicated` identifica si un valor de fila está duplicado. Las primeras apariciones también cuentan como duplicados.

```python
# coffees.select(pl.col("coffee_name").value_counts(sort=True)).unnest(pl.col("coffee_name"))
```

```python
# Filtra el DataFrame para mantener solo las filas donde el nombre del café
# aparece exactamente una vez en todo el conjunto de datos.
# En este dataset, es útil para encontrar pedidos especiales o únicos.
coffees.filter(pl.col("coffee_name").is_unique())
```

```python
# Su función es identificar todas las filas donde el valor de la columna coffee_name
# aparece más de una vez en el conjunto de datos. A diferencia de otros métodos,
# is_duplicated() marca como true tanto la primera aparición como todas las repeticiones posteriores
coffees.filter(pl.col("coffee_name").is_duplicated())
```

- El método `is_first_distinct` identifica si una fila contiene la primera aparición de un valor, independientemente de si es duplicado o no.
- El método `is_last_distinct` identifica si una fila contiene la última aparición de un valor, independientemente de si es duplicado o no.

```python
# is_first_distinct() marca como true la primera vez que aparece cada valor en la columna coffee_name.

# Es una forma muy eficiente de obtener una muestra de 'uno de cada tipo' de café
# presente en tus datos, conservando la primera fila donde se encuentra cada nombre distinto.
coffees.filter(pl.col("coffee_name").is_first_distinct())
```

```python
# Al igual que is_first_distinct(), este método identifica una única aparición
# de cada valor en la columna coffee_name, pero con una diferencia clave:
# marca como true la última vez que aparece cada nombre de café en el DataFrame.

# En la práctica, esto te permite obtener una lista con 'el registro más reciente'
# (asumiendo que los datos están ordenados por tiempo) de cada tipo de café disponible.
coffees.filter(pl.col("coffee_name").is_last_distinct())
```

## El Método `remove`

- El método `remove` _excluye_ filas que satisfacen una expresión.
- El método `remove` realiza la lógica inversa del método `filter`.

## `When`, `Then`, `Otherwise`
Los métodos `when`, `then` y `otherwise` en Polars se utilizan para crear expresiones condicionales, similares a la estructura `if-elif-else` de Python o `CASE WHEN` en SQL.

Aquí tienes cómo funciona cada parte:

* **pl.when(condición)**: Evalúa una condición booleana (por ejemplo, `pl.col('precio') > 10`).
* **.then(valor)**: Define qué valor asignar a las filas donde la condición anterior es verdadera (true).
* **.otherwise(valor)**: Define un valor por defecto para todas las filas que no cumplieron ninguna de las condiciones anteriores. Si no se usa, Polars asignará `null` por defecto.

Esta estructura es muy potente porque permite encadenar múltiples `when().then()` para manejar diversos escenarios en una sola columna.

```python
coffees = pl.read_csv("coffee_sales.csv", try_parse_dates=True).sort("time_of_purchase")
coffees.head(2)
```

- Digamos que queremos proporcionar una descripción para cada café basada en el tipo de leche.
- Si la leche es "Oat", la descripción debería ser `"Creamy"`. De lo contrario, el valor de respaldo es `"Delicious"`.
- La función `pl.lit` devuelve una expresión que representa un valor literal (es decir, fijo, constante).
- Sin un valor de respaldo, Polars usa un valor faltante (`null`) para cualquier valor que no coincida.

```python
coffees.with_columns(
    pl.when(pl.col("milk_type") == "Oat")
    .then(pl.lit("Creamy"))
    .alias("description")
)
```

- Usa el método `otherwise` para proporcionar un valor de respaldo en lugar de `null`.

```python
# Usar espacios para que el código sea más lagible!
coffees.with_columns(
    pl.when(pl.col("milk_type") == "Oat")
      .then(pl.lit("Creamy"))
    .otherwise(pl.lit("Delicious"))
    .alias("description")
)
```

- Podemos encadenar múltiples invocaciones del método `when` en secuencia.
- Cada llamada al método `when` crea una condición separada para evaluar.
- El diseño es similar a una sentencia `switch case` en varios lenguajes de programación.

```python
coffees.with_columns(
    pl.when(pl.col("milk_type") == "Oat")
      .then(pl.lit("Creamy"))
    .when(pl.col("milk_type") == "Almond")
      .then(pl.lit("Watery"))
    .otherwise(pl.lit("Delicious"))
    .alias("description")
)
```

## Particionando DataFrames
- Supongamos que queremos extraer un `DataFrame` para cada café único (`"Flat White"`, `"Latte"`).

```python
coffees = pl.read_csv("coffee_sales.csv", try_parse_dates=True)
coffees.head(3)
```

- Usa el método `unique` para devolver una columna de valores únicos de una columna objetivo.
- Escribir 12 sentencias `filter` es verboso y no es escalable.

```python
coffees.select(pl.col("coffee_name").unique())
```

- El método `partition_by` usa los valores únicos de una columna para devolver una lista de `DataFrames`.
- El método `partition_by` acepta una cadena con el nombre de la columna (una expresión no es válida).
- La palabra __partición__ significa "dividir en partes".
- Si tenemos 12 cafés diferentes, entonces `partition_by` devolverá una lista de 12 `DataFrames`.

```python
coffees.partition_by("coffee_name")
```

- La lista no es particularmente útil porque los `DataFrames` son accesibles por posición de índice.
- Polars almacenará los `DataFrames` en el orden en que encuentra los valores únicos en el `DataFrame` original.
- En este ejemplo, "Flat White" viene primero, luego "Latte", luego "Macchiato", etc.
- El parámetro `as_dict` devuelve un diccionario donde las claves son tuplas con los valores únicos de la columna y los valores son los `DataFrames`.

```python
# Se animan a explicar por que las claves del diccionario son tuplas? Ej. ('Flat White',)
coffees_dict = coffees.partition_by("coffee_name", as_dict=True)
coffees_dict
```

- Usa corchetes para indexar el diccionario y proporciona una clave de tupla para acceder al `DataFrame` correspondiente.


```python
coffees_dict[("Mocha",)]

# coffees_dict[("Affogato",)]
```

- La razón por la que el diccionario almacena claves de tupla es para soportar particionamiento con múltiples claves.
- En su lugar, dividamos el `DataFrame` por cada combinación de café y leche.

```python
coffee_and_milk = coffees.partition_by(["coffee_name", "milk_type"], as_dict=True)
# coffee_and_milk
```

- Pasa una tupla con la entrada de café y leche para extraer el `DataFrame` objetivo.
- Un solo valor (por ejemplo, "Americano") producirá un error.

```python
coffee_and_milk[("Americano", "Soy")]
```
