## Conversión de Mayúsculas/Minúsculas
- Polars agrupa métodos relacionados bajo un atributo/accesor (un espacio de nombres).
- Los métodos de cadenas de texto tienen un espacio de nombres dedicado `str`.

```python
nyc = pl.read_csv("nyc.csv")
nyc.head(2)
```

- El método `str.to_lowercase` convierte todos los caracteres a minúsculas.
- El método `str.to_uppercase` convierte todos los caracteres a mayúsculas.
- El método `str.to_titlecase` convierte todos los caracteres a formato título (primera letra de cada palabra en mayúscula).

```python
nyc.select(
    pl.col("Borough").str.to_lowercase(),
    pl.col("Pay Basis").str.to_uppercase(),
    pl.col("Agency Name").str.to_titlecase(),
)
```

- Convertir todas las columnas de texto a formato título parece ideal para este dataset.
- Podemos pasar un tipo de dato de Polars a la función `pl.col` para apuntar a las columnas con ese tipo de dato.
- `with_columns` reemplazará elegantemente las columnas antiguas con las nuevas (si tienen el mismo nombre).

```python
nyc.with_columns(pl.col(pl.String).str.to_titlecase())
```

## Eliminación de Espacios en Blanco

```python
nyc = pl.read_csv("nyc.csv")
nyc.head(2)
```

- El método `str.strip_chars` elimina los espacios en blanco del inicio y el final de la cadena.
- El método `str.strip_chars_start` elimina los espacios en blanco del inicio de la cadena.
- El método `str.strip_chars_end` elimina los espacios en blanco del final de la cadena.

```python
nyc.select(
    pl.col("Agency Name"),
    pl.col("Agency Name").str.strip_chars().alias("Agency Name without Whitespace"),
)
```

- Podemos encadenar múltiples métodos de cadena.
- Cada llamada a un método requiere el atributo `str`. El método se encuentra _dentro_ del espacio de nombres.
- Primero eliminemos los espacios en blanco de todas las columnas de texto, luego convirtamos sus valores a formato título.

```python
nyc.with_columns(pl.col(pl.String).str.strip_chars().str.to_titlecase())
```

## Eliminación de Prefijo y Sufijo

```python
nyc = pl.read_csv("nyc.csv").with_columns(
    pl.col(pl.String).str.strip_chars().str.to_titlecase()
)
nyc.head(2)
```

- El método `str.strip_prefix` elimina una subcadena consistente del inicio de cada cadena.
- Polars eliminará el prefijo exactamente una vez.
- Por ejemplo, podemos eliminar `Office Of ` del inicio de cada cadena que incluya ese texto.

```python
nyc.select(
    pl.col("Agency Name"),
    pl.col("Agency Name").str.strip_prefix("Office Of ").alias("Agency"),
)
```

- El método complementario `str.strip_suffix` elimina una subcadena del final de cada cadena.

```python
nyc.select(
    pl.col("Agency Name"),
    pl.col("Agency Name")
    .str.strip_prefix("Office Of ")
    .str.strip_suffix(" Authority")
    .alias("Agency"),
)
```

## Segmentación de Cadenas (Slicing)

```python
nyc = pl.read_csv("nyc.csv").with_columns(
    pl.col(pl.String).str.strip_chars().str.to_titlecase()
)
nyc.head(2)
```

- El método `str.slice` extrae una secuencia de caracteres de cada cadena.
- El primer argumento es el índice inicial, y el segundo argumento es el número de caracteres a extraer.
- Asignemos a cada municipio un código de 3 letras basado en sus primeras 3 letras.

```python
nyc.select(
    pl.col("Borough")
    , pl.col("Borough").str.slice(0, 3).alias("Code")
    )
```

- Si el segmento se extiende más allá del final de la cadena, Polars recogerá todos los caracteres.

```python
nyc.select(pl.col("Borough"), pl.col("Borough").str.slice(0, 100).alias("Code"))
```

- Podemos comenzar en el medio de una cadena especificando un índice distinto de cero.

```python
nyc.select(pl.col("Borough"), pl.col("Borough").str.slice(2, 5).alias("Code"))
```

- El método `str.head` extrae caracteres desde el inicio de la cadena.
- El método `str.tail` extrae caracteres desde el final de la cadena.
- Al igual que con `str.slice`, si la cantidad solicitada se extiende más allá de la longitud de la cadena, Polars extraerá tantos caracteres como sea posible.

```python
nyc.select(
    pl.col("Borough"),
    pl.col("Borough").str.head(3).alias("Code"),
    pl.col("Borough").str.tail(4).alias("Code 2"),
)
```

## Métodos de Filtrado

```python
nyc = pl.read_csv("nyc.csv").with_columns(
    pl.col(pl.String).str.strip_chars().str.to_titlecase()
)
nyc.head(2)
```

- El método `str.contains` devuelve verdadero si una fila contiene una subcadena dada.

```python
nyc.filter(pl.col("Title").str.contains("Water"))
```

- El método `str.contains_any` devuelve verdadero si el valor de la fila se encuentra dentro de un conjunto de valores posibles.

```python
nyc.filter(pl.col("Title").str.contains_any(["Water", "Fire"]))
```

- El método `str.starts_with` verifica si una subcadena se encuentra al inicio de una cadena.
- El método `str.ends_with` verifica si una subcadena se encuentra al final de la cadena.

```python
# nyc.filter(pl.col("Title").str.starts_with("Fire"))

# nyc.filter(pl.col("Title").str.ends_with("Marshal"))
```

- La distinción entre mayúsculas y minúsculas puede ser un problema. No hay valores de fila que comiencen con "fire" en minúsculas.

```python
nyc.filter(pl.col("Title").str.starts_with("fire"))
```

- Una solución es normalizar el texto, convirtiendo todos los caracteres a un formato consistente.
- Por ejemplo, podemos convertir todos los caracteres a minúsculas y luego proporcionar una subcadena en minúsculas.
- Las expresiones regulares (patrones de búsqueda para texto) son una solución alternativa que cubriremos en la siguiente lección.

```python
nyc.filter(pl.col("Title").str.to_lowercase().str.starts_with("fire"))

nyc.filter(pl.col("Title").str.to_uppercase().str.starts_with("FIRE"))
```

## Expresiones Regulares
- Una expresión regular es un patrón de búsqueda para texto.
- Una expresión regular usa símbolos para designar un patrón de coincidencia (cualquier dígito, un carácter, un espacio, etc).
- Las expresiones regulares pueden construir patrones de búsqueda más complejos (por ejemplo, "exactamente 2 dígitos seguidos de un espacio y cualquier carácter entre 'a' y 'e'").

```python
nyc = pl.read_csv("nyc.csv").with_columns(
    pl.col(pl.String).str.strip_chars().str.to_titlecase()
)
nyc.head(2)
```

- La mayoría de los métodos de cadena bajo `str` admiten argumentos de expresiones regulares.
- Usa cadenas crudas (raw strings) para expresiones regulares. Una cadena cruda le dice a Python que interprete cada carácter literalmente.
- Por ejemplo, Python tratará `\n` como una barra invertida seguida de `n` en lugar de como un carácter de nueva línea.
- Las cadenas crudas previenen conflictos entre la sintaxis de expresiones regulares y los caracteres de escape de Python.
- El siguiente ejemplo usa regex para identificar cualquier cadena que contenga _algún_ dígito.

```python
nyc.filter(pl.col("Title").str.contains(r"\d")).select(
    pl.col("Title")
).to_series().to_list()[:10]
```

- El siguiente ejemplo encuentra las cadenas con al menos 2 espacios en blanco seguidos.

```python
nyc.filter(pl.col("Title").str.contains(r"\s{2,}")).select(
    pl.col("Title")
).to_series().to_list()[:10]
```

## Grupos de Captura
- Un grupo de captura es una porción designada de la expresión regular.
- Polars puede extraer un valor coincidente a través del nombre de su grupo de captura.

```python
nyc = pl.read_csv("nyc.csv").with_columns(
    pl.col(pl.String).str.strip_chars().str.to_titlecase()
)
nyc.head(2)
```

- El método `str.extract` extrae el contenido de un grupo de captura.
- Digamos que queremos identificar los varios _tipos_ de mecánicos que trabajan en la ciudad de Nueva York.
- Queremos extraer la _palabra_ antes de "mechanic" para las entradas en la columna `"Title"`.
- Definiremos esa palabra antes de "mechanic" como un grupo de captura.
- `(?i)` es un patrón que hace la búsqueda insensible a mayúsculas/minúsculas. Es un ejemplo de [inline flag](https://docs.python.org/3/library/re.html#flags).
- `\w+` es un patrón para "1 o más caracteres de palabra (`[A-Za-z0-9_]`)".
- El grupo de captura 0 será la coincidencia completa (los caracteres de palabra, el espacio y `mechanic`).

```python
nyc.select(
    pl.col("Title"),
    pl.col("Title")
    .str.extract(r"(?i)(\w+)\smechanic", group_index=0)
    .alias("mechanic_type"),
)
```

- El índice 1 capturará el primer y único grupo de captura.
- El grupo de captura es el contenido correspondiente a `\w+` (uno o más caracteres) dentro de los paréntesis.

```python
nyc.select(
    pl.col("Title"),
    pl.col("Title")
    .str.extract(r"(?i)(\w+)\smechanic", group_index=1)
    .alias("mechanic_type"),
).filter(pl.col("mechanic_type").is_not_null())
```

## Reemplazo de Valores

```python
nyc = pl.read_csv("nyc.csv").with_columns(
    pl.col(pl.String).str.strip_chars().str.to_titlecase()
)
nyc.head(2)
```

- El método `str.replace` reemplaza la primera ocurrencia de una cadena con otra cadena.

```python
nyc.select(
    pl.col("Agency Name"),
    pl.col("Agency Name")
    .str.replace("Office Of", "Department of")
    .alias("Agency Name 2"),
)
```

- El método `str.replace_all` reemplaza _todas_ las ocurrencias de la subcadena.

```python
nyc.select(
    pl.col("Agency Name"),
    pl.col("Agency Name").str.replace("Of", "!!").alias("replace"),
    pl.col("Agency Name").str.replace_all("Of", "!!").alias("replace_all"),
)
```

- Polars admite expresiones regulares como argumentos.

```python
nyc.select(
    pl.col("Agency Name"),
    pl.col("Agency Name").str.replace(r"(?i)em", "!!").alias("replace"),
    pl.col("Agency Name").str.replace_all(r"(?i)em", "!!").alias("replace_all"),
)
```
